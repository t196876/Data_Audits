***Step 1***

```python
from PyPDF2 import PdfReader, PdfWriter

input_pdf = "combined 23.pdf"
output_pdf = "rotated_combined23.pdf"

reader = PdfReader(input_pdf)
writer = PdfWriter()

for page in reader.pages:
    page.rotate(90)
    writer.add_page(page)

with open(output_pdf, "wb") as f:
    writer.write(f)

print("PDF Rotated and saved as:",output_pdf)

```

***Step 2***

```python
import pdfplumber
 
rotated_pdf = "rotated_combined23.pdf"
all_lines=[]

with pdfplumber.open(rotated_pdf) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            lines = [line.strip() for line in text.split('\n') if line.strip()]
            all_lines.extend(lines)

print("Lines fetched", len(all_lines))
print("\n".join(all_lines[:]))
```

***Step 3***

```python
#test
import pandas as pd
import re
# Assume all_lines is already extracted using pdfplumber (like in your image)
# Replace this with actual extracted lines from rotated_combined11.pdf
# Example:
# all_lines = ["255465835 MDAK01", "PAUL GALBREATH", "115 GREENVIEW CIR", "FAYETTEVILLE GA 30214-2515", ...]
# Initialize list to store records
records = []
i = 0
while i < len(all_lines):
   line = all_lines[i].strip()
   # Step 1: Check if line looks like a Household ID + Creative Version
   match = re.match(r"(\d{8,})\s+([A-Z0-9]+)", line)
   if match:
       household_id = match.group(1)
       creative_version = match.group(2)
       # Step 2: Get full name (next line)
       if i+1 < len(all_lines):
           full_name = all_lines[i+1].strip()
       else:
           full_name = ""
       # Step 3: Address lines (1-2 lines max)
       address_line1 = all_lines[i+2].strip() if i+2 < len(all_lines) else ""
       address_line2 = all_lines[i+3].strip() if i+3 < len(all_lines) else ""
       # Step 4: Combine address smartly
       if re.search(r"\d{5}", address_line2):  # line2 has ZIP, it's city-state-ZIP
           full_address = f"{address_line1}, {address_line2}"
           i += 4
       else:
           full_address = address_line1
           i += 3
       # Step 5: Save record
       records.append({
           "full_name": full_name,
           "household_id": household_id,
           "full_address": full_address,
           "creative_version": creative_version
       })
   else:
       i += 1  # Skip non-matching lines
# Convert to DataFrame
df = pd.DataFrame(records)
# Display
print("Final structured DataFrame with", len(df), "rows")
df
```

```python

import pandas as pd
import numpy as np
from difflib import SequenceMatcher

# Load the Excel files
df_audit = pd.read_excel("S20250570_Automation_Output.xls", sheet_name="Matches", header=[0, 1], engine="xlrd")
df_ocr = pd.read_excel("corrected_output.xlsx", engine="openpyxl")

# Flatten multi-level headers
df_audit.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_audit.columns]

# Rename audit columns for easier access
df_audit = df_audit.rename(columns={
    'full_name_Audit': 'full_name',
    'household_id_Audit': 'household_id',
    'full_address_Audit': 'full_address',
    'creative_version_Audit': 'creative_version'
})

# Helper functions
def is_name_match(name1, name2, threshold=0.8):
    return SequenceMatcher(None, str(name1).strip().upper(), str(name2).strip().upper()).ratio() >= threshold

def normalize_id(val):
    try:
        return str(int(float(val)))
    except:
        return str(val).strip()

def is_missing(val):
    val = str(val).strip().upper()
    return pd.isna(val) or val in ["MISSING", "", "NAN", "-", "- - -"]

# Apply comparison and correction logic
for _, row in df_ocr.iterrows():
    name_ocr = row['full_name']
    matched_idx = None

    for idx, name_audit in df_audit['full_name'].dropna().items():
        if is_name_match(name_ocr, name_audit):
            matched_idx = idx
            break

    if matched_idx is not None:
        for col in ['full_name', 'household_id', 'full_address', 'creative_version']:
            ocr_col = f"{col}_OCR"
            if ocr_col in df_audit.columns:
                audit_val = row[col]
                ocr_val = df_audit.at[matched_idx, ocr_col]

                if col == 'household_id':
                    audit_val = normalize_id(audit_val)
                    ocr_val = normalize_id(ocr_val)

                if is_missing(ocr_val) or str(audit_val).strip().upper() != str(ocr_val).strip().upper():
                    df_audit.at[matched_idx, ocr_col] = audit_val

# Save the updated Excel file
df_audit.to_excel("corrected_output_final_scripted.xlsx", index=False)
print("Corrected output saved as corrected_output_final_scripted.xlsx.")
```
